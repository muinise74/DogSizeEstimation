## Data Processing

```python
imgList = []
```

```python
animal.iloc[0]['image']
```

```python
for i in range(len(animal)):
    path = animal.iloc[i]['image']
    img = plt.imread('./img/'+path)
    imgList.append(img)
```

```python
imgList
```

```python
hList=[]
wList=[]
cList=[]
```

```python
for i in imgList:
    h,w,c = i.shape
    hList.append(h)
    wList.append(w)
    cList.append(c)
```

```python
animal['height']=hList
animal['width']=wList
```

```python
del animal['nothing']
del animal['stand']
del animal['sit']
del animal['lie']
del animal['go']
del animal['run']
```

```python
del animal['cat']
del animal['cow']
del animal['horse']
del animal['sheep']
```

```python
imgList = []
dataList = []
for i in range(len(animal)):
    img_path = 'img/'+animal.iloc[i]['image']
    img_width=animal.iloc[i]['width']
    img_height=animal.iloc[i]['height'] 
    joint = [[animal.iloc[i]['L_eye_x'],animal.iloc[i]['L_eye_y'],animal.iloc[i]['L_eye_v']],[animal.iloc[i]['R_eye_x'],animal.iloc[i]['R_eye_y'],animal.iloc[i]['R_eye_v']],[animal.iloc[i]['L_ear_x'],animal.iloc[i]['L_ear_y'],animal.iloc[i]['L_ear_v']],[animal.iloc[i]['R_ear_x'],animal.iloc[i]['R_ear_y'],animal.iloc[i]['R_ear_v']],[animal.iloc[i]['Nose_x'],animal.iloc[i]['Nose_y'],animal.iloc[i]['Nose_v']],[animal.iloc[i]['Throat_x'],animal.iloc[i]['Throat_y'],animal.iloc[i]['Throat_v']],[animal.iloc[i]['Tail_x'],animal.iloc[i]['Tail_y'],animal.iloc[i]['Tail_v']],[animal.iloc[i]['withers_x'],animal.iloc[i]['withers_y'],animal.iloc[i]['withers_v']],[animal.iloc[i]['L_F_elbow_x'],animal.iloc[i]['L_F_elbow_y'],animal.iloc[i]['L_F_elbow_v']],[animal.iloc[i]['R_F_elbow_x'],animal.iloc[i]['R_F_elbow_y'],animal.iloc[i]['R_F_elbow_v']],[animal.iloc[i]['L_B_elbow_x'],animal.iloc[i]['L_B_elbow_y'],animal.iloc[i]['L_B_elbow_v']],[animal.iloc[i]['R_B_elbow_x'],animal.iloc[i]['R_B_elbow_y'],animal.iloc[i]['R_B_elbow_v']],[animal.iloc[i]['L_F_knee_x'],animal.iloc[i]['L_F_knee_y'],animal.iloc[i]['L_F_knee_v']],[animal.iloc[i]['R_F_knee_x'],animal.iloc[i]['R_F_knee_y'],animal.iloc[i]['R_F_knee_v']],[animal.iloc[i]['L_B_knee_x'],animal.iloc[i]['L_B_knee_y'],animal.iloc[i]['L_B_knee_v']],[animal.iloc[i]['R_B_knee_x'],animal.iloc[i]['R_B_knee_y'],animal.iloc[i]['R_B_knee_v']],[animal.iloc[i]['L_F_paw_x'],animal.iloc[i]['L_F_paw_y'],animal.iloc[i]['L_F_paw_v']],[animal.iloc[i]['R_F_paw_x'],animal.iloc[i]['R_F_paw_y'],animal.iloc[i]['R_F_paw_v']],[animal.iloc[i]['L_B_paw_x'],animal.iloc[i]['L_B_paw_y'],animal.iloc[i]['L_B_paw_v']],[animal.iloc[i]['R_B_paw_x'],animal.iloc[i]['R_B_paw_y'],animal.iloc[i]['R_B_paw_v']]]
    data ={'img_bbox':[animal.iloc[i]['bounds_x'],animal.iloc[i]['bounds_y'],animal.iloc[i]['bounds_width'],animal.iloc[i]['bounds_height']],'img_width':img_width,'img_height':img_height,'img_path':img_path,'joints':joint}
    imgList.append(img_path)
    dataList.append(data)
```

```python
imgList,dataList
```

```python
json_animal = {}
```

```python
for i in range(len(animal)):
    json_animal[imgList[i]]=dataList[i]
```

```python
for i in imgList:
    json_animal[i]['img_bbox'][0] = int(json_animal[i]['img_bbox'][0])
    json_animal[i]['img_bbox'][1] = int(json_animal[i]['img_bbox'][1])
    json_animal[i]['img_bbox'][2] = int(json_animal[i]['img_bbox'][2])
    json_animal[i]['img_bbox'][3] = int(json_animal[i]['img_bbox'][3])
    for j in range(20):
        json_animal[i]['joints'][j][0] = int(json_animal[i]['joints'][j][0])
        json_animal[i]['joints'][j][1] = int(json_animal[i]['joints'][j][1])
        json_animal[i]['joints'][j][2] = int(json_animal[i]['joints'][j][2])
        json_animal[i]['img_width'] = int(json_animal[i]['img_width'])
        json_animal[i]['img_height'] = int(json_animal[i]['img_height'])
        
        print(type(json_animal[i]['joints'][j][2]))
```
