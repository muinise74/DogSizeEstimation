```python
imgList = []
dataList = []
for i in range(len(animal)):
    img_path = 'img/'+animal.iloc[i]['image']
    joint = [[animal.iloc[i]['L_eye_x'],animal.iloc[i]['L_eye_y'],animal.iloc[i]['R_eye_v']],[animal.iloc[i]['R_eye_x'],animal.iloc[i]['R_eye_y'],animal.iloc[i]['R_eye_v']],[animal.iloc[i]['L_ear_x'],animal.iloc[i]['L_ear_y'],animal.iloc[i]['L_ear_v']],[animal.iloc[i]['R_ear_x'],animal.iloc[i]['R_ear_y'],animal.iloc[i]['R_ear_v']],[animal.iloc[i]['Nose_x'],animal.iloc[i]['Nose_y'],animal.iloc[i]['Nose_v']],[animal.iloc[i]['Throat_x'],animal.iloc[i]['Throat_y'],animal.iloc[i]['Throat_v']],[animal.iloc[i]['Tail_x'],animal.iloc[i]['Tail_y'],animal.iloc[i]['Tail_v']],[animal.iloc[i]['withers_x'],animal.iloc[i]['withers_y'],animal.iloc[i]['withers_v']],[animal.iloc[i]['L_F_elbow_x'],animal.iloc[i]['L_F_elbow_y'],animal.iloc[i]['L_F_elbow_v']],[animal.iloc[i]['R_F_elbow_x'],animal.iloc[i]['R_F_elbow_y'],animal.iloc[i]['R_F_elbow_v']],[animal.iloc[i]['L_B_elbow_x'],animal.iloc[i]['L_B_elbow_y'],animal.iloc[i]['L_B_elbow_v']],[animal.iloc[i]['R_B_elbow_x'],animal.iloc[i]['R_B_elbow_y'],animal.iloc[i]['R_B_elbow_v']],[animal.iloc[i]['L_F_knee_x'],animal.iloc[i]['L_F_knee_y'],animal.iloc[i]['L_F_knee_v']],[animal.iloc[i]['R_F_knee_x'],animal.iloc[i]['R_F_knee_y'],animal.iloc[i]['R_F_knee_v']],[animal.iloc[i]['L_B_knee_x'],animal.iloc[i]['L_B_knee_y'],animal.iloc[i]['L_B_knee_v']],[animal.iloc[i]['R_B_knee_x'],animal.iloc[i]['R_B_knee_y'],animal.iloc[i]['R_B_knee_v']],[animal.iloc[i]['L_F_paw_x'],animal.iloc[i]['L_F_paw_y'],animal.iloc[i]['L_F_paw_v']],[animal.iloc[i]['R_F_paw_x'],animal.iloc[i]['R_F_paw_y'],animal.iloc[i]['R_F_paw_v']],[animal.iloc[i]['L_B_paw_x'],animal.iloc[i]['L_B_paw_y'],animal.iloc[i]['L_B_paw_v']],[animal.iloc[i]['R_B_paw_x'],animal.iloc[i]['R_B_paw_y'],animal.iloc[i]['R_B_paw_v']]]
    data ={'img_bbox':[animal.iloc[i]['bounds_x'],animal.iloc[i]['bounds_y'],animal.iloc[i]['bounds_width'],animal.iloc[i]['bounds_height']],'img_path':img_path,'joints':joint}
    imgList.append(img_path)
    dataList.append(data)
```
